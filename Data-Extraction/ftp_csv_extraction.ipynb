```python
from ftplib import FTP
ftp = FTP('ftpcluster.loopia.se')
ftp.login("username","password")
files = ftp.nlst()

newfile = files[-1]
```

```python
date = newfile.split(" - ")[-1].split(".")[0] + ' 2017'

import datetime
newdate = datetime.datetime.strptime(date,'%b %d %Y').strftime('%Y-%m-%d')

dl_csv = newdate + '.csv'
```

```python
import pandas as pd
df = pd.read_csv(dl_csv, delimiter=";", encoding='cp1252')

df["date"]= pd.Series([newdate]*len(df))

df["Region"] = df["Region"].str.replace(u"ร", "A")
df.columns = df.columns.str.replace(u"รถ","o")
```
